#### Top 15 Happiest Countries

It's important to be able to visualize the data for the highest ranking countries. So, for each year, I will create a bar chart that shows the scores of the top 15 countries each year.

```python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(20,20))
ax1 = fig.add_subplot(221)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(223)
ax4 = fig.add_subplot(224)


top_countries_2019 = total_years['2019'].sort_values('Overall rank',ascending=True).iloc[:15]
top_countries_2018 = total_years['2018'].sort_values('Overall rank',ascending=True).iloc[:15]
top_countries_2017 = total_years['2017'].sort_values('Overall rank',ascending=True).iloc[:15]
top_countries_2016 = total_years['2016'].sort_values('Overall rank',ascending=True).iloc[:15]

#top_countries.plot.bar('Country or region','Score')

ax1.bar(top_countries_2019['Country or region'],top_countries_2019['Score'],color='green',edgecolor="black")
ax1.set_title('2019 Top 15 Happiest Countries')
ax1.set_xlabel('Country or Region')
ax1.set_ylabel('Overall Hapiness Score')

ax2.bar(top_countries_2018['Country or region'],top_countries_2018['Score'],color='blue',edgecolor="black")
ax2.set_title('2018 Top 15 Happiest Countries')
ax2.set_xlabel('Country or Region')
ax2.set_ylabel('Overall Hapiness Score')


ax3.bar(top_countries_2017['Country or region'],top_countries_2017['Score'],color='red',edgecolor="black")
ax3.set_title('2017 Top 15 Happiest Countries')
ax3.set_xlabel('Country or Region')
ax3.set_ylabel('Overall Hapiness Score')

ax4.bar(top_countries_2016['Country or region'],top_countries_2016['Score'],color='orange',edgecolor="black")
ax4.set_title('2016 Top 15 Happiest Countries')
ax4.set_xlabel('Country or Region')
ax4.set_ylabel('Overall Hapiness Score')

ax1.tick_params(labelrotation=45)
ax2.tick_params(labelrotation=45)
ax3.tick_params(labelrotation=45)
ax4.tick_params(labelrotation=45)

plt.show()
```

As we can see from the data, the reoccuring countries are consistent.
Finland, Denmark, Norway, Iceland, Switzerland, and the Netherlands are consistently in the top 5 spots throughout all 4 years.
The rest of the countries differ yearly, but contain Sweden, New Zealand, Israel.

An interesting observation is that in 2016 and 2017, the United States was in the top 15 spots, but in 2018 and 2019 the United States dropped out. That may be due to poltical tensions, and general unhappiness that has occured throughout the media over political issues in America.
